```python
import pandas as pd

data = pd.read_csv('stocks.csv', index_col='Symbol')

print(data)

print(data.info())

print(f"Number of rows: {data.shape[0]}, Number of columns: {data.shape[1]}")

print(data.head(10))
print(data.tail(10))

print(data['Company'])

print(data[['Company', 'Industry']])

print(data.loc['AAPL'])

print(data.loc['AAPL', 'Exchange'])

sort = data.sort_values(by=['Industry', 'Company'])
print(sort)

Name_data = data[data['Industry'] == 'Information Technology']
print(Name_data)

note_data = data[data['Notes'].notnull()]
print(note_data)


no_notes = data.drop(columns=['Notes'])
print(no_notes)

price_data = pd.read_csv('Stock-Prices.csv')


merged_data = data.join(price_data.set_index('Symbol'), on='Symbol')
print(merged_data)
```
